##Pozitif Korelasyonlar

```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots(2,2,figsize=(15,10))
sns.scatterplot(x='cevre',y='radius_worst',data=df,hue='tani',ax=ax[0][0])
sns.scatterplot(x='alan',y='radius_worst',data=df,hue='tani',ax=ax[1][0])
sns.scatterplot(x='doku',y='texture_worst',data=df,hue='tani',ax=ax[0][1])
sns.scatterplot(x='area_worst',y='radius_worst',data=df,hue='tani',ax=ax[1][1])
plt.show()
```

##Negatif Korelasyonlar

```python
fig,ax=plt.subplots(2,2,figsize=(15,10))
sns.scatterplot(x='alan',y='fraktal_boyut',data=df,hue='tani',ax=ax[0][0])
sns.scatterplot(x='yaricap',y='smoothness_se',data=df,hue='tani',ax=ax[1][0])
sns.scatterplot(x='smoothness_se',y='cevre',data=df,hue='tani',ax=ax[0][1])
sns.scatterplot(x='alan',y='smoothness_se',data=df,hue='tani',ax=ax[1][1])
```

```python
scatter = px.scatter(df,x='yaricap',y='cevre',color='tani',size_max=60)
scatter.show()
```

```python
selected_features = df.iloc[:,2:]

# Korelasyon matrisi ve ısı haritası
correlation_matrix = selected_features.corr()
plt.figure(figsize=(17, 14), dpi=100)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Özellikler Arasındaki Korelasyon')
plt.show()
```

##Kategorik verileri Binary formatına dönüştürme

```python
from sklearn.preprocessing import LabelEncoder

column_to_encode = 'tani'

label_encoder = LabelEncoder()

df[column_to_encode] = label_encoder.fit_transform(df[column_to_encode])

df.to_csv('encoded_data.csv', index=False)
df.head(5)
```
